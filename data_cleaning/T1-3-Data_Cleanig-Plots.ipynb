## Scatter Plot

```python
df.plot(kind="scatter", x="Day Mins", y= "Day Charge")  # linear correlation - linear regression
```

```python
df.plot(kind="scatter", x="Night Mins", y= "Night Charge")
```

```python
figure, axs = plt.subplots(2,2, sharey=True, sharex=True)
df.plot(kind="scatter", x="Day Mins", y="Day Charge", ax= axs[0][0]) # subplot
df.plot(kind="scatter", x="Night Mins", y="Night Charge", ax= axs[0][1])
df.plot(kind="scatter", x="Day Calls", y="Day Charge", ax= axs[1][0])
df.plot(kind="scatter", x="Night Calls", y="Night Charge", ax= axs[1][1])
```

## Frequency Histogram

```python
plt.hist(df["Day Calls"], bins=20) # divisions  [0,30,60,90,120,150]
plt.xlabel("Number of calls per day")
plt.ylabel("Frequency of calls per day")
plt.title("Histogram of number of calls per day")
```

```python
k = int(np.ceil(1+np.log2(3333)))  # Sturges Rule
plt.hist(df["Day Calls"], bins= k)
plt.xlabel("Number of calls per day")
plt.ylabel("Frequency of calls per day")
plt.title("Histogram of number of calls per day")
```

## BoxPlot - Box-and-whisker plot

```python
plt.boxplot(df["Day Calls"])
plt.ylabel("Number of calls per day")
plt.title("Boxplot Day Calls")
```
